## these data are already clustered but we try k-means to cluster them then make comaprison

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Create synthetic data
X, y_true = make_blobs(n_samples=500, centers=5, cluster_std=0.60, random_state=0)

# Step 2: Apply KMeans clustering
kmeans = KMeans(n_clusters=5, random_state=42)
y_pred = kmeans.fit_predict(X)

# Step 3: Plot both original and KMeans clustering
plt.figure(figsize=(12, 5))

# Plot 1: True labels (original clusters)
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis')
plt.title("Original Clusters (from make_blobs)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)

# Plot 2: KMeans predicted clusters
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='x', label='Centroids')
plt.title("KMeans Predicted Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=1000, noise=0.05)
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
dbscan.components_  #This gives you the actual core points  is equal to X[dbscan.core_sample_indices_]
```

```python
dbscan.core_sample_indices_  #Indices of core points (into your original X)
```

### the dbscan has no predict function so we use KNN

```python
# the dbscan just classify the existing data but cant predict new instance so knn takes these data and predict the new data
# dbscan make labels for features (make it classification) then knn treained on these data and predict the new instance 
from sklearn.neighbors import KNeighborsClassifier

# Step 1: Fit KNN on DBSCAN's core points
knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(dbscan.components_, dbscan.labels_[dbscan.core_sample_indices_])

# Step 2: Get new data points
X_new = np.array([[-0.5, 0], [0, 0.5], [1, -0.1], [2, 1]])

# Step 3: Compute distance to nearest neighbor (but no prediction yet)
y_dist, y_pred_idx = knn.kneighbors(X_new, n_neighbors=1)

#  Step 4: Now manually get the predicted label from core sample labels
y_pred = dbscan.labels_[dbscan.core_sample_indices_][y_pred_idx]

# Step 5: Mark far points as outliers
y_pred[y_dist > 0.2] = -1

# Step 6: Final result
print(y_pred.ravel())

```
